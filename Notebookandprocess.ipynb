```python
import os

import pandas as pd

AWS_S3_BUCKET = os.getenv("AWS_S3_BUCKET")
AWS_ACCESS_KEY_ID = os.getenv("AWS_ACCESS_KEY_ID")
AWS_SECRET_ACCESS_KEY = os.getenv("AWS_SECRET_ACCESS_KEY")
AWS_SESSION_TOKEN = os.getenv("AWS_SESSION_TOKEN")

key = "geomapping/inputdata/devicestatus.txt/devicestatus.txt"

devicestatus_df = pd.read_csv(
    f"s3a://geomapping/inputdata/devicestatus.txt",
    storage_options={
        "key": AWS_ACCESS_KEY_ID,
        "secret": AWS_SECRET_ACCESS_KEY,
        "token": AWS_SESSION_TOKEN,
    },sep='[,|]',usecols=[0,1,2,12,13],header=None,names=['date','models','deviceid','Latitude','Longitude',]
)

```

```python
devicestatus_df.dropna(inplace=True)
```

```python
devicestatus_df[['Manufacturer','Model_Name']]=devicestatus_df['models'].str.rsplit(n=1,expand=True)
```

```python
devicestatus_df.insert(1, 'Manufacturer', devicestatus_df.pop('Manufacturer'))
```

```python
devicestatus_df.insert(2, 'Model_Name', devicestatus_df.pop('Model_Name'))
```

```python
devicestatus_df.head()
```

```python
devicestatus_df=devicestatus_df.drop(columns='models')
```

```python
devicestatus_df.to_csv(
    f"s3a://geomapping/outputfolder/devicestatus.txt",
    index=False,
    storage_options={
        "key": AWS_ACCESS_KEY_ID,
        "secret": AWS_SECRET_ACCESS_KEY,
        "token": AWS_SESSION_TOKEN,
    },
)
```
